```python
#Inspecting Ticket_Price_Cr column before cleaning
df_raw = pd.read_csv("../data/raw/Luxury_Housing_Bangalore.csv")
df_raw['Ticket_Price_Cr'].str.contains('Cr', na=False).sum()
df_raw[df_raw['Ticket_Price_Cr'].str.contains('Cr', na=False)].head(5)
```

```python
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].astype(str)

# Remove ₹ symbol
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].str.replace('₹', '', regex=False)

# Remove 'Cr'
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].str.replace('Cr', '', regex=False)

# Remove spaces
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].str.strip()

# Convert to numeric
df['Ticket_Price_Cr'] = pd.to_numeric(df['Ticket_Price_Cr'], errors='coerce')
```

```python
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].fillna(df['Ticket_Price_Cr'].median())
```

```python
df['Unit_Size_Sqft'] = pd.to_numeric(df['Unit_Size_Sqft'], errors='coerce')
```

```python
cols = ['Micro_Market','Developer_Name','Configuration',
        'Transaction_Type','Buyer_Type','Sales_Channel','Possession_Status']

for col in cols:
    df[col] = df[col].str.strip().str.lower()
```

```python
df['Price_per_Sqft'] = (df['Ticket_Price_Cr'] * 10000000) / df['Unit_Size_Sqft']
```

```python
df["Micro_Market"] = df["Micro_Market"].str.title()
df["Developer_Name"] = df["Developer_Name"].str.title()
```

```python
df["Transaction_Type"].value_counts(dropna=False)
```

```python
df["Transaction_Type"] = df["Transaction_Type"].astype(str).str.strip().str.title()

df["Booking_Flag"] = df["Transaction_Type"].apply(
    lambda x: 1 if x == "Primary" else 0
)

df["Booking_Status"] = df["Booking_Flag"].map({
    1: "Booked",
    0: "Not Booked"
})
```

```python
df["Purchase_Quarter"] = pd.to_datetime(df["Purchase_Quarter"])
df["Quarter_Number"] = df["Purchase_Quarter"].dt.quarter
```

```python
# Validate
print("Minimum value before cleaning:", df["Ticket_Price_Cr"].min())
print("Invalid rows:", (df["Ticket_Price_Cr"] <= 0).sum())
```

```python
# Remove invalid negative price rows
df = df[df["Ticket_Price_Cr"] > 0]

# Verify
print("Min price after cleaning:", df["Ticket_Price_Cr"].min())

# Save cleaned dataset
df.to_csv("cleaned_luxury_housing.csv", index=False)
```
